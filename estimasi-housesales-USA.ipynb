# Harga Rumah Berdasarkan bedrooms

```python
models = df.groupby('date').count()[['bedrooms']].sort_values(by='bedrooms',ascending=True).reset_index()
models = models.rename(
    columns={'bedrooms':'NumberOfdate'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['date'], y=models['NumberOfdate'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi harga rumah

```python
plt.figure(figsize=(15,5))
sns.distplot(df['bedrooms'])
```

# seleksi fitur

```python
features = ['bathrooms','bedrooms','waterfront','sqft_living','grade','sqft_above','yr_built','lat','long','zipcode','sqft_lot15','sqft_living15']
x = df[features]
y = df['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#bathrooms=1, #bedrooms=3, waterfront=0, sqft_living=1180, grade=7, sqft_above=1180, yr_built=1955, lat=47.5112, long=-122.257, zipcode=98178 sqft_lot15=5650,sqft_living15=1340
input_data = np.array([[1,3,0,1180,7,1180,1955,47.5112,-122.257,98178,5650,1340]])

prediction = lr.predict(input_data)
print('Estimasi Harga Rumah di King County, USA :', prediction)
```
